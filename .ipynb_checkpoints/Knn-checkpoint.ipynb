Import important libraries

```python
import random
import numpy as np
import pandas as pd
from collections import Counter
```

Read dataset using pandas

```python
df=pd.read_csv('knn_200_dataset.csv')
df.head()
```

```python
df.shape
```

Train-Test-split the dataset

```python
def train_test(x,y,split_size=0.2,random_state=2):
    random.seed(random_state)
    data=list(zip(x,y))
    random.shuffle(data)
    xs,ys=zip(*data)
    split=int(len(xs)*(1-split_size))
    xtr=xs[:split]
    xte=xs[split:]
    ytr=ys[:split]
    yte=ys[split:]
    return np.array(xtr),np.array(xte),np.array(ytr),np.array(yte)

x=df.drop(columns=['Label']).values
y=df['Label'].values

x_train,x_test,y_train,y_test=train_test(x,y,split_size=0.2,random_state=42)
```

Implementing Knn from scratch

```python
class KNN:
    def __init__(self,k=3):
        self.k=k

    def euclidean(self,x1,x2):
        return np.sqrt(np.sum((x1-x2)**2))

    def fit(self,x,y):
        self.x_train=x
        self.y_train=y

    def predict(self,x):
        prediction=[]
        for i in x:
            prediction.append(self.__predict(i))
        return np.array(prediction)

    def __predict(self,x):
        distance=[]
        for i in self.x_train:
            distance.append(self.euclidean(i,x))

        k_indices=np.argsort(distance)[:self.k]
        k_label=[self.y_train[i] for i in k_indices]
        most_common=Counter(k_label).most_common(1)
        return most_common[0][0]

    def accuracy(self,y_test,y_pred):
        total=len(y_test)
        correct=0
        for i in range(total):
            if(y_test[i]==y_pred[i]):
                correct+=1
        return (correct/total)*100
```

```python
knn=KNN()
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
print(pred)
```

Accuracy of the mode

```python
acc=knn.accuracy(y_test,pred)
print("Accuracy is",acc)
```
